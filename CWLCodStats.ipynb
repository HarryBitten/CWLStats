```python
import math 
  
# Function to calculate the Probability 
def Probability(rating1, rating2): 
  
    return 1.0 * 1.0 / (1 + 1.0 * math.pow(10, 1.0 * (rating1 - rating2) / 400)) 
  
  
# Function to calculate Elo rating 
# K is a constant. 
# d determines whether 
# Team A wins or Player B.  
def EloRating(Rwinteam, Rloseteam): 

    K=40
    d=1
    # To calculate the Winning 
    # Probability of Team B 
    Pb = Probability(Rwinteam, Rloseteam) 
  
    # To calculate the Winning 
    # Probability of Team A 
    Pa = Probability(Rloseteam, Rwinteam) 
  
    # Case -1 When Team A wins 
    # Updating the Elo Ratings 
    if (d == 1) : 
        Rwinteam = Rwinteam + K * (1 - Pa) 
        Rloseteam = Rloseteam + K * (0 - Pb) 
      
  
    # Case -2 When Team B wins 
    # Updating the Elo Ratings 
    else : 
        Rwinteam = Rloseteam + K * (0 - Pa) 
        Rwinteam = Rloseteam + K * (1 - Pb) 
      
  
    #print("Updated Ratings:-") 
    #print("Rwinteam =", round(Rwinteam, 6)," Rloseteam =", round(Rloseteam, 6)) 
    #print Rwinteam
    #print Rloseteam
    return Rwinteam, Rloseteam
  
```
